```python
import matplotlib.pyplot as plt 
import pickle
import numpy as np
from typing import List
```

```python
with open('pickle.pkl', 'rb') as f:
    data_dict = pickle.load(f)
data_dict.keys()
```

```python
annualReports = data_dict['annualReports']
annualReports
```

```python
totalRevenue = []
ebitda = []
fiscalDateEnding = []

for i in annualReports:
    totalRevenue.append(i['totalRevenue'])
    ebitda.append(i['ebitda'])
    fiscalDateEnding.append(i['fiscalDateEnding'])

ebitda = ebitda[::-1]
fiscalDateEnding = fiscalDateEnding[::-1]

revenue = []
for i in totalRevenue:
   revenue.append(int(i))
```

```python
max(revenue)
```

```python
plt.bar(fiscalDateEnding,revenue, color='orange',width=0.4)
plt.xlabel("Fiscal Date Ending")
plt.ylabel("Total Revenue")
plt.plot()
```
